#Zadanie 3
Z tej samej ramki danych wyświetl dane wykorzystując filtry oparte o mapowanie:

a) tylko kobiety z wykształceniem 'Master' oraz 'Bachelor' i pożyczkami pomiędzy 50000 a 70000,

b) wszyscy mężczyźni z wykształceniem innym niż 'Master' oraz będący właścicielem lokalu (kolumna person_home_ownership = OWN),

c) wszystkie osoby z wartością loan_percent_income w przedziale od 0.4 do 0.6,

d) wszystkie wartości pożyczek, których kwota jest wielokrotnością liczby 10000.

```python
#a) tylko kobiety z wykształceniem 'Master' oraz 'Bachelor' i pożyczkami pomiędzy 50000 a 70000,(nie ma rekordu)
map_1 = df['person_gender'] == 'female'
map_2 = df['person_education'] == 'Master'
map_3 = df['person_education'] == 'Bachelor'
map_4 = df['loan_amount'] >= 50000
map_5 = df['loan_amount'] <= 70000
df[map_1 & map_2 & map_3 & map_4 & map_5]
```

```python
# b) wszyscy mężczyźni z wykształceniem innym niż 'Master' oraz będący właścicielem lokalu (kolumna person_home_ownership = OWN),
map_1 = df['person_gender'] == 'male'
map_2 = df['person_education'] != 'Master'
map_3 = df['person_home_ownership'] == 'OWN'
df[map_1 & map_2 & map_3]
```

```python
# c) wszystkie osoby z wartością loan_percent_income w przedziale od 0.4 do 0.6,
map_1 = df['loan_percent_income'] >= 0.4
map_2 = df['loan_percent_income'] <= 0.6
df[map_1 & map_2]
```

```python
#d) wszystkie wartości pożyczek, których kwota jest wielokrotnością liczby 10000.

df[df['loan_amount'] % 1000 == 0]
```
